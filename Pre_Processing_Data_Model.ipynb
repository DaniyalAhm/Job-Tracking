```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split



encoder = OneHotEncoder(handle_unknown='ignore', min_frequency=5)

df = pd.read_csv('Final_Augmented_Training_Data.csv')

max_len =(max([len(i) for i in df["Body"]]))
df.head()
#Had to do manuel encoding because the onehotencoder was did not have frequency encoding
discovered_words = set()
padded  = []
for text in df['Body']:
    words = text.split()
    for word in words:
        discovered_words.add(word)  

df['Tokens'] = df['Body'].str.split()

new_df = pd.DataFrame()
for word in discovered_words:
    new_df[word] = df['Tokens'].apply(lambda x: x.count(word))

X_Encoded = new_df.to_numpy()
print(X_Encoded)
```

```python
Bayes_classifier = MultinomialNB()
X_train, X_test, y_train, y_test = train_test_split(X_Encoded, df['Jobs'], test_size=0.2, random_state=42)

Bayes_classifier.fit(X_train, y_train)
print(Bayes_classifier.score(X_test, y_test))


```

```python
from sklearn.metrics import f1_score
y_pred = Bayes_classifier.predict(X_test)
print(f1_score(y_test, y_pred, average='weighted'))
```
