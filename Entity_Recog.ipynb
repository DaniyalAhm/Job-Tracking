```python
import pandas as pd 

df = pd.read_csv("Filtered_Mail.csv")
df = df[df["Date"].str.contains("2024")]
df= df[~df["From"].str.contains("<messages-noreply@linkedin.com> ")]
df["Body"] = df["Body"].apply(lambda x: x.replace("\n", ""))
df["Body"] = df["Body"].apply(lambda x: x.replace("\r", ""))
df["Body"] = df["Body"].apply(lambda x: x.replace("\t", ""))
df["Body"] = df["Body"].apply(lambda x: x.replace("-", ""))
df["Body"].head(10)
```

```python
import nltk
nltk.download('all')           # Tokenizer
       # Word list
```

```python
from nltk import sent_tokenize, word_tokenize, ne_chunk, pos_tag


sentence_tokenizer = lambda x: sent_tokenize(x)
word_tokenizer = lambda sentences: [word_tokenize(sentence) for sentence in sentences]


tokens = [sentence_tokenizer(x) for x in df["Body"]]
pos_tags = [pos_tag(token) for token in tokens]
print(pos_tags)
```

```python
ner_tree = [ne_chunk(sentence) for sentence in pos_tags]
```

```python
def extract_entities(tree):
    entities = []
    for subtree in tree:
        if isinstance(subtree, nltk.Tree):
            entity = " ".join([token for token, pos in subtree.leaves()])
            entity_type = subtree.label()
            entities.append((entity, entity_type))
    return entities


entities = [extract_entities(tree) for tree in ner_tree]
```
